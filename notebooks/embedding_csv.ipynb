```python
import pandas as pd
import os
import csv
import numpy as np
from ast import literal_eval
```

```python
embedding_csv = '/home/steffi/dev/independent_study/fairface_margin_025/embeddings/val.csv'
```

```python
fair_face_embeddings = pd.read_csv(embedding_csv, quoting=csv.QUOTE_ALL, delimiter=',')
```

```python
fair_face_embeddings["embedding"] = fair_face_embeddings["Embedding"].apply(lambda embedding: np.array(','.join(','.join(embedding[2:-2].split(' \n')).split()).split(','), dtype=float))
```

```python
fair_face_embeddings["embedding"]
```

```python
fair_face_embeddings["file"] = fair_face_embeddings["Image Path"].apply(lambda im_path: '/'.join(im_path.split('/')[-2:]))
```

```python
fair_face_embeddings.drop(columns=["Image Path", "Embedding"], inplace=True)
```

```python
fair_face_embeddings.head()
```

# Joing Dataframes: FairFace train labels and face embeddings

```python
labels_csv = '/home/steffi/dev/independent_study/fairface_margin_025/fairface_label_val.csv'
```

```python
labels = pd.read_csv(labels_csv)
```

```python
labels.head()
```

```python
merged = pd.merge(labels, fair_face_embeddings, how='inner', on='file', sort=False)
```

```python
merged.head()
```

```python
merged_csv = '/home/steffi/dev/independent_study/fairface_margin_025/embeddings/val_labels_and_embeddings.csv'
merged.to_csv(merged_csv, index=False)
```
