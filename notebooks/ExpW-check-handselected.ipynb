```python
import pandas as pd
import os
import csv
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def get_file_names(dir_name):
    file_names = []
    for root, dirs, files in os.walk(dir_name, topdown = False):
        for name in files:
            new_name = "_".join(name.split("_")[:-1]) + ".png"
            file_names.append(new_name)
    return file_names
```
